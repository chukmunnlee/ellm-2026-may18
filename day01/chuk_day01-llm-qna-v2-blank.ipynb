## DistilBERT base cased distilled SQuAD
DistilBERT is a small, fast, cheap and light Transformer model trained by distilling BERT base. More details [here](https://huggingface.co/distilbert/distilbert-base-cased-distilled-squad).

```python
model_name = "distilbert/distilbert-base-cased-distilled-squad"
```

```python
# TODO: Create a tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# TODO: Encode text
message = 'hello, world'
message = 'The polar bear is a large bear native to the Arctic and nearby areas'
message = 'thou shalt not kill'

encoded_msg = tokenizer(message, return_tensors='pt')
print(encoded_msg)

for tok in encoded_msg['input_ids'][0]:
    text = tokenizer.decode(tok)
    print(f'{tok} = {text}')

decoded_msg = tokenizer.decode(encoded_msg['input_ids'][0], skip_special_tokens=True)
print(decoded_msg)

```

## Working with LLMs
Create and instance of the Large Language Model (LLM). We will then create a simple
prompt, tokenize the prompt and feed the tokenized prompt to the LLM. The response
from the LLM will be decoded to human friendly text.

```python
# TODO: Load libraries
from transformers import AutoModelForQuestionAnswering
import torch
```

```python
# load the model, same model as the tokenizer
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
# TODO: Load question answer model
idx = 1000

context = dataset['train'][idx]['story_structure']
context = dataset['train'][idx]['infilled_story']
question = dataset['train'][idx]['question']
expected_answer = dataset['train'][idx]['expected_answer']

print(f'Context: {context}')
print(f'Question: {question}')
print(f'Expected answer: {expected_answer}')
```

```python
# TODO: Encode context and question
input = tokenizer(question, context, return_tensors='pt', padding=True, truncation=True)
print(input)
decode_input = tokenizer.decode(input['input_ids'][0])
print(decode_input)
```

```python
# TODO: Tokenize the inputs
result = model(input['input_ids'], input['attention_mask'])
print(result.keys())
```

```python
print(result['start_logits'])
print('------------------------')
print(result['end_logits'])
```

```python
start_pos = torch.argmax(result['start_logits'])
end_pos = torch.argmax(result['end_logits']) + 1

print(f'Start position: {start_pos}')
print(f'End position: {end_pos}')

enc_ans = input['input_ids'][0][start_pos:end_pos]
print(enc_ans)

ans = tokenizer.decode(enc_ans)
print(f'Answer: {ans}')
print(f'Expected answer: {expected_answer}')

```
